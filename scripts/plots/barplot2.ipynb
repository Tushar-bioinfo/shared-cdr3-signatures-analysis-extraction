```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Corrected data
data = [
    {"Cancer": "Uterus", "Kmer": "4-mer", "Significant": 8},
    {"Cancer": "Uterus", "Kmer": "5-mer", "Significant": 5},
    {"Cancer": "Kidney", "Kmer": "4-mer", "Significant": 0},
    {"Cancer": "Kidney", "Kmer": "5-mer", "Significant": 0},
    {"Cancer": "Lung_Adeno", "Kmer": "4-mer", "Significant": 1},
    {"Cancer": "Lung_Adeno", "Kmer": "5-mer", "Significant": 1},
    {"Cancer": "Lung_SCC", "Kmer": "4-mer", "Significant": 0},
    {"Cancer": "Lung_SCC", "Kmer": "5-mer", "Significant": 0},
]
df_counts = pd.DataFrame(data)

# Fix palette to match Kmer values
palette = {
    "4-mer": "#4bcffa",  # light blue
    "5-mer": "#ef5777"   # green
}

# Plot
plt.figure(figsize=(8, 5))
sns.barplot(data=df_counts, x="Cancer", y="Significant", hue="Kmer", palette=palette)
plt.title("Number of Significant TRB Motifs by Cancer Type and K-mer")
plt.ylabel("Motifs with p < 0.05")
plt.xlabel("Cancer Type")
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig("fig3_significant_motifs_per_cancer.png", dpi=300)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Original data
data = [
    {"Cancer": "Uterus", "Kmer": "4-mer", "Significant": 8},
    {"Cancer": "Uterus", "Kmer": "5-mer", "Significant": 5},
    {"Cancer": "Kidney", "Kmer": "4-mer", "Significant": 0},
    {"Cancer": "Kidney", "Kmer": "5-mer", "Significant": 0},
    {"Cancer": "Lung_Adeno", "Kmer": "4-mer", "Significant": 1},
    {"Cancer": "Lung_Adeno", "Kmer": "5-mer", "Significant": 1},
    {"Cancer": "Lung_SCC", "Kmer": "4-mer", "Significant": 0},
    {"Cancer": "Lung_SCC", "Kmer": "5-mer", "Significant": 0},
]
df = pd.DataFrame(data)

# Group non-uterus cancers into one group
df["Cancer"] = df["Cancer"].apply(lambda x: "Uterus" if x == "Uterus" else "Other_Cancers")

# Aggregate counts
df_grouped = df.groupby(["Cancer", "Kmer"], as_index=False).sum()

# Custom color palette
palette = {
    "4-mer": "#ef5777",  # blue
    "5-mer": "#4bcffa"   # green
}

# Plotting
plt.figure(figsize=(6, 5))
sns.barplot(data=df_grouped, x="Cancer", y="Significant", hue="Kmer", palette=palette, edgecolor="black", linewidth=1.5,alpha=0.8)
plt.title("Significant TRB Motifs: Uterus vs Other Cancers (p < 0.05)")
plt.ylabel("Number of Significant Motifs")
plt.xlabel("Cancer Type")
plt.tight_layout()
plt.savefig("fig3_uterus_vs_others.png", dpi=300)
plt.show()
```
